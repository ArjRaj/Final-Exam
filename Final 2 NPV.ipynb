### 1. Create a Numpy array of size 20 elements using random package. Convert the above numpy array into 2 equal arrays of size 2x5 each. Convert the second array to a 1-D array and reverse the elements.And find the 4th element in the reversed array.(5 Marks)

```python
import numpy as np
import pandas as pd

import random
arr = np.random.randint(0,20,20)
print(arr)
```

```python
temp = np.split(arr,2)
arr1 = temp[0].reshape(2,5)
arr2 = temp[1].reshape(2,5)
print("Array 1:\n",arr1)
print("Array 2:\n",arr2) #Two equal arrays of size 2x5
```

```python
arr2_1 = arr2.ravel()
print(arr2_1) #Second array converted to 1D array
```

```python
arr2_1_r = arr2_1[::-1]
print(arr2_1_r) #Reverse of elements from the above array
```

```python
print(arr2_1_r[3]) #Fetching 4th element in the Reversed array
```

### 4.	Write a program to 

A.	Create a pandas Series with consecutive numbers between 10 to 20 (include both the boundary values in the series). (1 Mark)

B.	Replace all the even numbers with number 30. (1 Mark)

C.	Display the series in descending order. (1 Mark)

D.	Create a Data frame from the above-created series with column names as numbers. Fetch the data from 2nd, 6th, and 7th rows. 
(2 Marks)


```python
#A)

a = np.arange(10,21)
s = pd.Series(a)
print(s)
```

```python
#B)

for i in s.index:
    if s[i]%2 == 0:
        s[i] = 30
        
print("Series after replacing the Even numbers with 30:\n",s)
```

```python
#C)
s_desc = s.sort_values(ascending=False).reset_index(drop=True)
print("The Series in descending order:\n",s_desc)
```

```python
#D)

df1 = pd.DataFrame(s_desc)
df1.columns = ['numbers']
print("The Data Frame is:\n\n",df1)
```
