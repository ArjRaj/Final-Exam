```python
df3=pd.read_csv("C:/Users/Anand/Desktop/Yash/Fifa.csv")
df3.head()
```

#### 1. Data Manipulation for Value Column (5 Marks).
Step1: Clean the Value column by removing &#39;?&#39; from each row.
Step2: Convert the column in uniform unit i.e. &#39;K&#39; by using given conversion rate 1 =1000 K and also
remove the letter &#39;K&#39; and &#39;M&#39; from each row.
Step3: Convert the Datatype of the column to &#39;float&#39;
First 5 values of Sample output column should look like this:

0      750.0
1    21500.0
2     4000.0

3      240.0
4      150.0
Name: Value, dtype: float64

```python
df3['Value']=df3['Value'].str.replace('?','')
df3['Value']=df3['Value'].str.replace('K','')
df3['Value']=df3['Value'].str.replace('.','')
df3['Value']=df3['Value'].str.replace('M','000')
df3['Value']=df3['Value'].astype(float)

df3['Value'].head()
```

#### 2. Data Manipulation for Height Column (5 Marks).
Step1: Convert Height column from Feet’ Inches unit to Inches unit by using a given conversion rate
1 feet = 12 inches 
Step2: Convert the Datatype of the Height column to &#39;int&#39;
First 5 values of Sample Output Height column should look like this:

0  61
1   67
2 67
3 61
4 66
Name: Height, dtype: int32

```python
lis1_height = []
for i in df3.Height:
  i = int(i[0])*12+ int(i[2])
  lis1_height.append(i)
    

df3['height']= lis1_height
df3['height'].head()
```

#### 3. Data Manipulation for Weight Column (5 Marks).
Step1: Clean Weight column by removing ‘lbs’ from each row’
Step2: Convert the Datatype of the Weight column to &#39;int&#39;
First 5 values of Sample Output Weight column should look like this:

0    143
1    150
2    159
3    168
4    154
Name: Weight, dtype: int32

```python
weight = []
for i in df3.Weight.str[0:3]:
  weight.append(float(i))
df3['Weight'] = weight
df3['Weight'].head()
```
