#a)

We go for PCA in order to reduce the curse of dimentionality and when there are huge amount of multicollinearity in the data.

```python
#B) PCA for the whole data
from sklearn.decomposition import PCA

pca = PCA()
df_pca = pca.fit_transform(df_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
cov_matrix = np.cov(df_scaled.T)
print('Covariance Matrix \n%s', cov_matrix)
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
print('Eigen Vectors \n%s', eig_vecs)
print('\n Eigen Values \n%s', eig_vals)
```

```python
eigen_pairs = [(np.abs(eig_vals[i]), eig_vecs[ :, i]) for i in range(len(eig_vals))]
```

```python
tot = sum(eig_vals)
var_exp = [( i /tot ) * 100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
print("Cumulative Variance Explained", cum_var_exp)
```

Number of PCA Components:

    From the above we can see that 20 PCA Components explain the 90% variance of the data, hence we use 20 PCA components 
    for further use.

```python
plt.figure(figsize=(6 , 4))
plt.bar(range(29), var_exp, alpha = 0.5, align = 'center', label = 'Individual explained variance')
plt.step(range(29), cum_var_exp, where='mid', label = 'Cumulative explained variance')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Components')
plt.legend(loc = 'best')
plt.tight_layout()
plt.show()
```

```python
#Outliers Treatment:

def outliers(df,col):
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    low_lim=q1-(1.5*iqr)
    up_lim=q3+(1.5*iqr)
    
    for x in df[col]:
        if x>up_lim:
            df[col]=df[col].replace(to_replace=x,value=up_lim)
        elif x<low_lim:
            df[col]=df[col].replace(to_replace=x,value=low_lim)
```

```python
for i in list(pca_df.columns):
    outliers(pca_df,i)
```

```python
#a)
# Kmeans:
```

```python
from sklearn.cluster import KMeans
cluster_range = range( 1, 20 )
cluster_errors = []
for num_clusters in cluster_range:
    clusters = KMeans( num_clusters, n_init = 10,random_state=0)
    clusters.fit(pca_df)
    cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
clusters_df
```

```python
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

```python
kmeans = KMeans(n_clusters=8, n_init = 15, random_state=3)
```

```python
kmeans.fit(pca_df)
```

```python
centroids = kmeans.cluster_centers_
```

```python
centroid_df = pd.DataFrame(centroids, columns = list(df_final.columns) )

centroid_df
```

```python
kmeans.labels_
```

```python
pca_dfcopy=pca_df.copy()
pca_dfcopy['Class']=list(kmeans.labels_)
pca_dfcopy.head()
```

```python
pca_dfcopy['Class'].value_counts()
```

```python
#Agglomerative Clustering
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram 
plt.figure(figsize=[10,10])
merg = linkage(pca_df, method='ward')
dendrogram(merg, leaf_rotation=90)
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hie_clus = AgglomerativeClustering(n_clusters=15, affinity='euclidean', linkage='ward')
cluster2 = hie_clus.fit_predict(pca_df)

pca_df_agg = pca_df.copy(deep=True)
pca_df_agg['label'] = cluster2
pca_df_agg
```

#b)
Optimal Number of cluster in K-Means are 8 and Agglomerative Clustering is 15.

#c)
Comparison of K-means and Agglomerative Clustering using shilhouette Score:

```python
from sklearn.metrics import silhouette_score

silhouette_k = silhouette_score(pca_df, pca_dfcopy['Class'])
silhouette_h = silhouette_score(pca_df, pca_df_agg['label'])
print('Silhouette score for K-Means Clistering',silhouette_k)
print('Silhouette score for Agglomerative Clistering',silhouette_h)
```

Comparison of K-means and Agglomerative Clustering using Inertia Score:

```python
print('Inertia of K-Means :',kmeans.inertia_)

print('Inertia of Agglomerative Clustering :',kmeans.inertia_)
```

We can see that the Silhoutte score for K-Means is better than Agglomerative clustering, hence we continue with Agglomerative
