### 5.	Split dataset into train and test (70:30) (5 marks)

a.	Are both train and test representative of the overall data? How would you ascertain this statistically?


```python
X = df_final.drop('Yearly Amount Spent',axis=1)
y = df_final['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

Since the target variable is a continuous variable, we perfom One-way ANOVA to ascertain this statistically.

Hence we state the Hypothesis as follows,

H0: Both train and test represent the overall data.

H1: Both train and test do not represent the overall data.

```python
from scipy.stats import f_oneway
f_oneway(y_train,y_test,y)
```

Inference:

1. Since p-value(0.611) is greater than alpha(0.05), we accept the Null Hypothesis.

2. Hence, we can say that both train and test are representative of the overall data.

```python
#A)
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
print("Overall Score for train:",lr.score(X_train,y_train))
print("Overall Score for test:",lr.score(X_test,y_test))
```

Inference: The model is performing good on both the train and test data.

```python
from sklearn.metrics import mean_squared_error,r2_score
y_pred = lr.predict(X_test)
print("R_2 Score:",r2_score(y_test,y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
#B)

import statsmodels.api as sm

Xc = sm.add_constant(X)
lr_ols = sm.OLS(y,Xc).fit()
lr_ols.summary()
```

Inference: 
1. The Adj. R2 = 0.984 and it is not different from R2 = 0.984.

2. In general, R2 is the ratio of sum of explained variance and total variance. Whereas, Adj. R2 is the ratio of R2 and statistical fluke.

3. Here, R2 and Adj. R2 are same since there are less number of features and every feature contributes to the model.

## Assumptions

1. No Autocorrelation

Durbin-Watnson = 1.887 which is almost 2. Hence there is no Auto Correlation

2. Normality of Residuals

Jarque-Bera(JB) = 0.198 < 5.99. Hence the Residuals are normally distributed.

3. Linearity Test

```python
print(sm.stats.diagnostic.linear_rainbow(lr_ols))
```

p-value(0.12)>alpha(0.05) which proves Linearity of Residuals.

4. Homoscedasticity Test

```python
import statsmodels.stats.api as sns
from statsmodels.compat import lzip
name=['F-statistic','p-value']
test = sns.het_goldfeldquandt(lr_ols.resid,lr_ols.model.exog)
lzip(name,test)
```

p-value(0.285)>alpha(0.05) which proves Homoscedasticity.

#C)

1. Based on the OLS model, the features having p-value less than alpha(0.05) are significant features.

2. We can observe that 'Time on Website' has p-value of 0.326 > alpha(0.05). Hence, 'Time on Website' is the insignificant feature.

```python
#D)

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=[variance_inflation_factor(Xc.values,i) for i in range(Xc.shape[1])]
pd.DataFrame({'VIF':vif[1:]},index=X.columns)
```

Inference: There is no Multi Collinearity as the VIF values are below 5.

### 7.	How do you improve the accuracy of the model? Write clearly the changes that you will make before re-fitting the model. Fit the final model. (20 marks)
Please feel free to have any number of iterations to get to the final answer. Marks are awarded based on the quality of final model you are able to achieve. 


```python
#Backward Elimination
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p=[]
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p=pd.Series(model.pvalues.values[1:],index=cols)
    pmax=max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

Inference: 
1. From the Backward Elimination, we can see that 'Time on Website' is not selected.

2. Hence we refit the final Linear model using the above selected features and remove the 'Time on Website' feature.

```python
X_BE = df_final[['Avg. Session Length','Time on App','Length of Membership']]
y_BE = df_final['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_BE,y_BE,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

lr_be = LinearRegression()
lr_be.fit(X_train,y_train)
print("Overall Score for train:",lr_be.score(X_train,y_train))
print("Overall Score for test:",lr_be.score(X_test,y_test))
```

```python
import statsmodels.api as sm

Xc = sm.add_constant(X_BE)
lr_ols = sm.OLS(y_BE,Xc).fit()
lr_ols.summary()
```

```python
y_pred = lr_be.predict(X_test)
from sklearn.metrics import mean_absolute_error
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,y_pred)
print("R_2 Score:",r2_score(y_test,y_pred))

print("Mean Squared Error:",mse)
print("Root Mean Squared Error:",rmse)
print("Mean Absolute Error:",mae)
```
